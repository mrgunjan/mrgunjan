```python
import numpy as np       #import library/module Numpy & Pandas
import pandas as pd
```

# **making data frame from csv file**

```python
pd.read_csv('/content/prac2.csv') #import csv file/readd

```

```python
a = pd.read_csv('/content/prac2.csv') #consider as ---- a = pd.read_csv
```

```python
b = a.fillna(0) #fill up the NaN/na Value as a 0
b
```

# **ffill() function is used forward fill the missing value in the dataframe**

```python
c = a.fillna(method= 'ffill')  #up side value put the up side (forword filed)
c
```

# bfill() is used to backward fill the missing values in the dataset. It will backward fill the NaN values that are present in the pandas dataframe.

```python
d = a.fillna(method='bfill') #down side value put the down side (backword filed)
d
```

```python
e = a.fillna(method='ffill',axis='columns') #to display the value forward column in table (Changes 5,6, Column)
e
```

```python
f = a.fillna(method='bfill',axis='columns')  #to display the value Backword column in table
f
```

```python
g = a.fillna(method='ffill',limit=2)  #to the limit forward fill with Only One Stage
g
```

```python
h = a.fillna(method='bfill',limit=3)  #to the limit Backward fill with Only Two Stage
h
```

```python
i = a.dropna() #this method Clear All NaN/na Value & Show Only Value Added Rows/Column
i #Pandas dropna() method allows the user to analyze and drop Rows/Columns with Null values.
```
